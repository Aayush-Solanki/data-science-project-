```python
from sklearn.cluster import KMeans

x = data.drop(['label'], axis=1)

x = x.values

print(x.shape)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

plt.rcParams['figure.figsize'] = (10, 4)

wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    km.fit(x)
    wcss.append(km.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method', fontsize=20)
plt.xlabel('No. Of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Assuming you have already defined and fitted your KMeans model (km) as in your previous code
km = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_mean = km.fit_predict(x)

a = data['label']
y_mean = pd.DataFrame(y_mean)  # Store the cluster assignments in a DataFrame

z = pd.concat([y_mean, a], axis=1)
z = z.rename(columns={0: 'clusters'})

print("Let's check the Results After Applying KMeans Clustering Analysis \n")
print("Crops in First Cluster:", z[z['clusters'] == 0]['label'].unique())
print("-----------------------------------")
print("Crops in Second Cluster:", z[z['clusters'] == 1]['label'].unique())
print("-----------------------------------")
print("Crops in Third Cluster:", z[z['clusters'] == 2]['label'].unique())
print("-----------------------------------")
print("Crops in Fourth Cluster:", z[z['clusters'] == 3]['label'].unique())
```

```python
y = data['label']
x = data.drop(['label'], axis = 1)

print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

print("The Shape of x trian:", x_train.shape)
print("The Shape of x test:", x_test.shape)
print("The Shape of y trian:", y_train.shape)
print("The Shape of y test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model
model = LogisticRegression(max_iter = 1000, solver = 'saga', C = 0.01)

# Fit the model to the training data
model.fit(x_train, y_train)

# Make predictions on the test data
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10, 10)

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Create a heatmap of the confusion matrix
sns.heatmap(cm, annot=True, cmap='Wistia')
plt.title("Confusion Matrix for Logistic Regression", fontsize=15)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

```python
from sklearn.metrics import classification_report

cr = classification_report(y_test, y_pred)
print(cr)
```
